```python
'''### Task 1: Evaluate AI Categorization Accuracy

Using the bookkeeper annotations as ground truth:

- Compare AI-generated `AssignedCategory` with the corrected `Category`
- Calculate precision, recall, F1 score for categorization (both overall and per-category if possible)
- Identify categories with highest and lowest accuracy

ðŸ“Ž *Bonus*: Visualize confusion between similar categories (e.g. â€œChair Rental:Hair Chairâ€ vs â€œChair Rental:Lash Bedâ€)'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support

csv_file_path = 'Categorized_Transactions_Elle_Jan2025(in).csv'

# Load the dataset
df = pd.read_csv(csv_file_path)

# Replace empty strings with NaN in the bookkeeper correction column ("Category")
df['BookkeeperCategory'] = df['Category'].replace('', np.nan)

# Filter rows where a bookkeeper correction is provided (i.e. BookkeeperCategory is not null)
df_corr = df.dropna(subset=['BookkeeperCategory'])

if df_corr.empty:
    print("No bookkeeper corrections available for evaluation.")
else:
    # Ground truth and prediction
    y_true = df_corr['BookkeeperCategory']
    y_pred = df_corr['AssignedCategory']
    
    # Overall classification report
    print("Classification Report:")
    print(classification_report(y_true, y_pred, zero_division=0))
    
    # Overall precision, recall, and F1 score (weighted)
    overall_precision, overall_recall, overall_f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average='weighted', zero_division=0)
    print("Overall Metrics:")
    print(f"Precision: {overall_precision:.2f}")
    print(f"Recall: {overall_recall:.2f}")
    print(f"F1 Score: {overall_f1:.2f}")
    
    # Calculate per-category accuracy
    category_accuracy = {}
    for category in y_true.unique():
        mask = (y_true == category)
        if mask.sum() > 0:
            correct = (y_pred[mask] == category).sum()
            category_accuracy[category] = correct / mask.sum()
    
    sorted_accuracy = sorted(category_accuracy.items(), key=lambda x: x[1], reverse=True)
    print("\nPer-category Accuracy:")
    for cat, acc in sorted_accuracy:
        print(f"{cat}: {acc:.2f}")
    
    if sorted_accuracy:
        highest = sorted_accuracy[0]
        lowest = sorted_accuracy[-1]
        print(f"\nHighest Accuracy Category: {highest[0]} with accuracy {highest[1]:.2f}")
        print(f"Lowest Accuracy Category: {lowest[0]} with accuracy {lowest[1]:.2f}")
    
    # Visualize the confusion matrix
    labels = np.unique(np.concatenate([y_true.unique(), y_pred.unique()]))
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    
    plt.figure(figsize=(12,10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel("Assigned Category")
    plt.ylabel("Bookkeeper Category")
    plt.title("Confusion Matrix")
    plt.show()

#f"Direct Deposit Payable, Chair Rental:Facial Room, Chair Rental:Hair Chair, 60400 Bank Service Charges, Cash Clearing Account, Chair Rental:Lash Bed, 67100 Rent Expense, Cost of Goods Sold:Salon Supplies:Supplies and Materials, 68600 Utilities, 51800 Merchant Account Fees, Uncategorized Expense, 20000 Accounts Payable, 24000 Payroll Liabilities:Federal Taxes, 25550 PST Payable (BC):PST Payable (BC), TD Visa 1968, Uncategorized Income, 25500 GST/HST Payable:GST/HST Payable, 66000 Payroll Expenses:WCB Expenses, 64300 Meals and Entertainment, Subcontractor, Sales of Product Income, 15000 Furniture and Equipment, Other Income, 61700 Computer and Software  Expenses, Sales, 30800 Owners Draw, Cost of Labor:Commission, Legal & Professional Fess:Professional Fees - Accounting, Rent Security Deposit, Undeposited Funds, MyJobs_test, Bank loans
```

```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Categorized_Transactions_Elle_Jan2025(in).csv')

# Step 1: Create ground truth category
data['ground_truth_category'] = data['Category'].fillna(data['AssignedCategory'])

# Step 2: Calculate overall accuracy
# Accuracy is the proportion of transactions where bookkeeper 'Category' is NaN (AI correct)
overall_accuracy = data['Category'].isna().mean()
print(f"Overall Accuracy: {overall_accuracy:.4f}")

# Step 3: Compute per-category metrics
true_labels = data['ground_truth_category']
predicted_labels = data['AssignedCategory']
report = classification_report(true_labels, predicted_labels, output_dict=True)
report_df = pd.DataFrame(report).transpose()

# Filter to category-specific metrics (exclude 'accuracy', 'macro avg', 'weighted avg')
categories = data['ground_truth_category'].unique()
category_report = report_df.loc[categories]

print(categories)
print(category_report)

# Identify categories with highest and lowest F1 scores
highest_f1_category = category_report['f1-score'].idxmax()
lowest_f1_category = category_report['f1-score'].idxmin()
print(f"Category with Highest F1-Score: {highest_f1_category} (F1: {category_report.loc[highest_f1_category, 'f1-score']:.4f})")
print(f"Category with Lowest F1-Score: {lowest_f1_category} (F1: {category_report.loc[lowest_f1_category, 'f1-score']:.4f})")

# Step 4: Bonus - Confusion Matrix
cm = confusion_matrix(true_labels, predicted_labels, labels=categories)
plt.figure(figsize=(16, 14))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=categories, yticklabels=categories)
plt.xlabel('Predicted Category')
plt.ylabel('True Category')
plt.title('Confusion Matrix of AI Categorizations')
plt.tight_layout()
plt.show()
```

To ensure the categorization system remains robust over time, itâ€™s essential to implement an evaluation framework that:

Periodically Monitors Metrics: Regularly computes key performance metrics (e.g., overall accuracy, precision, recall, F1-score, confusion matrices) on a validation set or live stream of incoming data.

Tracks Performance Drift: Compares current metrics with historical benchmarks to detect significant changes that might indicate drift.

2. Minimal Evaluation Pipeline
Below is an outline (and optional code snippet) for a pipeline that leverages incoming bookkeeper feedback:

Pipeline Steps:
Data Collection:

Live Data Ingestion: Capture all transactions along with their AI-assigned categories and subsequent bookkeeper corrections.

Store Feedback: Maintain a log (or database table) that tracks each transactionâ€™s ground truth label, AI prediction, and a ConfidenceScore (if available).

Metric Computation:

Batch Evaluation: Periodically (e.g., daily or weekly) run evaluations on a batch of transactions.

Compute Metrics: Use standard metrics such as accuracy, precision, recall, and F1-score. Additionally, generate confusion matrices to identify systematic misclassifications.

Drift Detection:

Historical Comparison: Compare the current metrics with historical performance. Use statistical tests (e.g., a moving average or control charts) to flag significant deviations.

Alert Mechanism: Set thresholds for acceptable performance levels. If the metrics fall outside these thresholds, trigger an alert for further investigation.

Flagging Uncertain Predictions:

Confidence-Based Fallback: Use the ConfidenceScore from the model to flag predictions that fall below a defined threshold.

Human Review Queue: Route these uncertain predictions to a human review queue for manual verification.

Incorporating Human Corrections:

Feedback Loop: Automatically incorporate corrections from bookkeepers into the training dataset.

Periodic Retraining: Schedule retraining sessions (e.g., monthly or when drift is detected) that include the new, corrected examples to update the model.

```python
# Optional Code Outline for the Evaluation Pipeline

import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def load_recent_feedback():
    return pd.read_csv('recent_feedback.csv')

def compute_metrics(df):
    # Ensure that the 'ground_truth_category' is set based on bookkeeper corrections
    df['ground_truth_category'] = df['Category'].fillna(df['AssignedCategory'])
    true_labels = df['ground_truth_category']
    predicted_labels = df['AssignedCategory']
    report = classification_report(true_labels, predicted_labels, output_dict=True, zero_division=0)
    cm = confusion_matrix(true_labels, predicted_labels)
    return report, cm

def plot_confusion_matrix(cm, labels):
    plt.figure(figsize=(12,10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.title("Confusion Matrix")
    plt.ylabel("True Labels")
    plt.xlabel("Predicted Labels")
    plt.tight_layout()
    plt.show()

def detect_drift(current_metrics, historical_metrics, threshold=0.05):
    # Compare current F1-score with historical average, and flag if deviation exceeds threshold
    drift_flags = {}
    for category in current_metrics:
        if category in historical_metrics:
            diff = abs(current_metrics[category]['f1-score'] - historical_metrics[category]['f1-score'])
            drift_flags[category] = diff > threshold
    return drift_flags

def flag_uncertain_predictions(df, confidence_threshold=0.5):
    # Flag predictions with ConfidenceScore below the threshold
    df['flag_for_review'] = df['ConfidenceScore'] < confidence_threshold
    return df.loc[df['flag_for_review']]

def evaluation_pipeline():
    # Load recent feedback data
    data = load_recent_feedback()
    
    # Compute evaluation metrics
    report, cm = compute_metrics(data)
    print("Classification Report:\n", report)
    
    # Plot confusion matrix
    labels = sorted(data['ground_truth_category'].unique())
    plot_confusion_matrix(cm, labels)
    
    # Simulated historical metrics (e.g., loaded from previous runs or monitoring system)
    historical_metrics = {
        # Example structure: 'Category': {'f1-score': value, ...}
        'Direct Deposit Payable': {'f1-score': 1.0},
        'Research Expense': {'f1-score': 0.0},
        # ... add metrics for all categories as baseline
    }
    
    # Detect drift by comparing current and historical metrics
    drift_flags = detect_drift(report, historical_metrics, threshold=0.05)
    print("Drift Detection Flags by Category:", drift_flags)
    
    # Flag uncertain predictions based on ConfidenceScore
    uncertain_df = flag_uncertain_predictions(data, confidence_threshold=0.5)
    print("Transactions flagged for human review:")
    print(uncertain_df[['accountId', 'name', 'ConfidenceScore', 'AssignedCategory', 'ground_truth_category']])
    
    # Save uncertain predictions for manual review
    uncertain_df.to_csv('uncertain_predictions.csv', index=False)

    # At this point, human corrections can be integrated into the training dataset for future retraining.
    # The pipeline can be scheduled to run periodically.

if __name__ == "__main__":
    evaluation_pipeline()

```
